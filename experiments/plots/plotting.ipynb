```python
def plot(data):
    colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFD700', '#FF6347', '#8A2BE2', '#40E0D0', '#D2691E']  # 8 Colors for partitions
    labels = ['compress', '<=', 'rowMin', 'transpose','mm','binaryOps', 'other']  # Labels for partitions
    names = ["Uncompressed", "Compressed-Base", "Intermediate-Compression"][::-1]
    names = ["Compressed-Base" ,"w/out fusion", "w/ fusion"]
    
    fig, ax = plt.subplots(figsize=(6, 6))
    x_positions = np.arange(data.shape[0]) * 2  # Position of pillars

    
    for i in range(data.shape[0]):
        bottom = 0
        for j in range(data.shape[1]):
            ax.bar(x_positions[i], data[i, j], width=1, color=colors[j], bottom=bottom, edgecolor='black')
            bottom += data[i, j]
        # Display total height above each pillar
        total_height = np.sum(data[i])
        ax.text(x_positions[i], total_height + 0.5, "{:.2f}".format(float(total_height)), ha='center', fontsize=12, fontweight='bold')
    
    ax.set_xticks(x_positions)
    ax.set_xticklabels(names[:data.shape[0]])
    ax.set_ylabel('time in s')
    ax.set_title('Local')
    #ax.set_yscale('log')
    ax.set_ylim(0, 45)
    #ax.set_ylim(0, 120)  # Set maximum y-axis value to 15
    
    # Add legend
    handles = [plt.Rectangle((0, 0), 1, 1, color=colors[j]) for j in range(data.shape[1])]
    ax.legend(handles, labels, title='Operations')
    
    plt.show()
```
